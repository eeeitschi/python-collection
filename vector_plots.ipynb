```python
import matplotlib.pyplot as plt
import numpy as np

# Array of vectors (start and endpoint for each vector)
vector_array = np.array([[0, 0, 0, 1, 1, 1]])

X, Y, Z, U, V, W = zip(*vector_array)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W)
ax.set_xlim([0, 3])
ax.set_ylim([0, 3])
ax.set_zlim([0, 3])
plt.show()
```
