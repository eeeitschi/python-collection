```python
import numpy as np
import matplotlib.pyplot as plt
import math

unten = 0.1
oben = 1000

#Funktionsplotter
def f(x):
    #return np.tan(x**3)
    return x**(2.5)

def g(x):
    #return np.tan(x**3)
    return x**2 * np.log2(x)

#Generiere x-Werte von 0 - 6 mit 1000 Sttzstellen
x = np.linspace(unten, oben, num=1000)
y = f(x)
x2 = np.linspace(unten, oben, num=1000)
y2 = g(x)

plt.figure(figsize=(20,10))
plt.grid()
plt.plot(x, y, x2, y2)
plt.legend(['f(x)', 'g(x)'])
```
